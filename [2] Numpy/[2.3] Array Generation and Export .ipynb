# I. Generating Data With NumPy

NumPy possess a lot of ways to populate and create data, making it an incredibly versatile tool for numerical computing. In this section we're going to discuss the most common ways to populate an array with NumPy aside from using `numpy.array()`

## A. Filling an array with zeros: [`np.zeros`](https://numpy.org/doc/stable/reference/generated/numpy.zeros.html)

`np.zeros()` creates an array filled with zeros.

It takes a shape tuple as an argument, which specifies the dimensions of the array.

This function is often used to initialize an array before populating it with actual data.

- Generate a 1-dimensional array of zeros

```python
zeros_array = np.zeros(5)
print(zeros_array)
```

- Generate a 2-dimensional array of zeros

```python
zeros_matrix = np.zeros((3, 4))
print(zeros_matrix)
```

- Generate a 5-dimensional array of zeros

```python
zeros_fived = np.zeros((1, 2, 3, 4, 5))
print(zeros_fived)
```

## B. Filling an array with ones: [`np.ones`](https://numpy.org/doc/stable/reference/generated/numpy.ones.html)

`np.ones()` creates an array filled with ones, similar to `np.zeros()`, it also takes a shape tuple as an argument.

This function is useful when you want to initialize an array with a default value of 1.

- Generate a 1-dimensional array of ones

```python
ones_array = np.ones(3)
print(ones_array)
```

- Generate a 3-dimensional array of ones-

```python
ones_3d = np.ones((4, 2, 3))
print(ones_3d)
```

## C. Diagonal Matrix: [`np.diag`](https://numpy.org/doc/stable/reference/generated/numpy.diag.html)

`np.diag()` creates a diagonal array where the values are placed along the main diagonal and zeros elsewhere.

It can take either a 1-dimensional array or a scalar as an argument.

This function is commonly used to create matrices with specific diagonal values.

- Generate a 2-dimensional array with specified diagonal values

```python
diag_array = np.diag([1, 2, 3])
print(diag_array)
```

- Generate a 2-dimensional array with specified diagonal values

```python
diag_matrix = np.diag(np.linspace(1,10,10))
print(diag_matrix)
```

- Extracting the diagonal elements of the matrix

```python
diag_matrix = np.diag(np.linspace(1,9,9).reshape(3,3))
print(diag_matrix)
```

Notice that operating `np.diag` twice will revert to the original matrix

- Extracting the diagonal elements of the matrix

```python
diag_matrix = np.diag(np.diag(np.linspace(1,10,10)))
print(diag_matrix)
```

## D. Geometric Progression: [`np.geomspace()`](https://numpy.org/doc/stable/reference/generated/numpy.geomspace.html)

`np.geomspace()` generates a 1-D array with values in a geometric progression.

You can specify the start, end, and number of elements you want in the array.

The function evenly spaces the values on a logarithmic scale, resulting in a geometric progression.

- Creating a 1D array with 5 values in a geometric progression from 1 to 32

```python
arr = np.geomspace(1, 32, 5)
print(arr)
```

- Creating a 1D array with 4 values in a geometric progression from 1000 to 0.1

```python
arr = np.geomspace(1000, 0.1, 4)
print(arr)
```

## E. Arrays generated from function of indices: [`np.fromfunction()`](https://numpy.org/doc/stable/reference/generated/numpy.fromfunction.html)

`np.fromfunction()` creates an array by applying a function to each coordinate.

You provide a function that takes the **indices** of the array as arguments and returns the corresponding value.

The function is evaluated for each index of the array to populate it.

- Creating a 1D array with values from 0 to 4 using a lambda function

```python
arr = np.fromfunction(lambda i: i, (5,))
print(arr)
```

- Creating a 2D array with values from a custom function

```python
arr = np.fromfunction(lambda i, j: i + j, (5, 4))
print(arr)
```

- Creating a 2D array with values from a custom function with multiple operations

```python
arr = np.fromfunction(lambda x, y: x*y+x, (4, 7))
print(arr)
```

## F. Tiling ndarrays: [`np.tile`](https://numpy.org/doc/stable/reference/generated/numpy.tile.html)

`np.tile()` allows you to replicate a given array or pattern to create a larger array. It constructs a new array by repeating the provided input along specified dimensions.

You pass the input array or pattern, and a tuple specifying the number of repetitions in each dimension.

The resulting array has dimensions calculated based on the input array and the repetition parameters. This function is useful when you need to create an array with repeated patterns or extend the size of an existing array.

- Creating a 4x4 array by tiling a 2x2 array

```python
arr = np.array([[1, 2],
                [3, 4]])

tiled_arr = np.tile(arr, (2, 2))
print(tiled_arr)
```

- Creating a 4x2 array by tiling a 2-1x1 array

```python
arr = np.array([[1],
                [34]])

tiled_arr = np.tile(arr, (2, 2))
print(tiled_arr)
```

- Creating a 2x4 array by tiling a 1x2 array

```python
arr = np.array([1, 34])

tiled_arr = np.tile(arr, (2, 2))
print(tiled_arr)
```

## G. Matrix with Scalar: [`np.full`](https://numpy.org/doc/stable/reference/generated/numpy.full.html)

`np.full()` creates a new NumPy array filled with a specified scalar value.

It takes two parameters: the shape of the array specified as a tuple of integers, and the constant value that will populate the array.

- Creating a 1D array of size 5 filled with 7

```python
arr = np.full(5, 7)
print(arr)
```

- Creating a 2D array of size 4x3 filled with -1.5

```python
arr = np.full([4, 3], -1.5)
print(arr)
```

```python
arr = np.full([2, 5, 3], 69)
print(arr)
```

# II. Using NumPy Meshgrid


## A. Basics

The `np.meshgrid()` function is used to create coordinate matrices from coordinate vectors. Given two or more 1D arrays representing points along different dimensions, this function returns two or more 2D arrays representing the coordinate matrices.

These arrays can be used to evaluate points on a grid or for vectorized evaluations of functions on the grid.

- The first step is to create coordinate vectors

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])
```

- Then we use meshgrid to create coordinate matrices

```python
X, Y = np.meshgrid(x, y)

print(X)
print(Y)
```

One of the most useful applications of `np.meshgrid()` is their use in contour plots.

In this example, we want to visualize the function $f(x, y) = x^2 - y^2$ on a 2D grid. We use `np.linspace()` to initialize the point values and `np.meshgrid()` to initialize arrays, which returns two 2D arrays, $X$ and $Y$, representing the coordinates of the grid points.

For this example, the range of our meshgrid $x$ and $y$ are from $-5$ to $5$

```python
x_values = np.linspace(-5, 5, 100)
y_values = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x_values, y_values)
```

Next, we calculate the function values for each point on the grid using the formula $f(x, y) = x^2 - y^2$, and store the results in a 2D array called $Z$.

```python
Z = X**2 - Y**2
```

To visualize the function, we create a filled contour plot using `plt.contourf()`. The plot displays points that represent contours of equal function values.

We add a color bar to the plot to indicate the value of $f(x, y)$ for each color. The x and y axes are labeled, and the plot is given a title.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, Z, cmap='hsv', levels=100)
plt.colorbar(label='$f(x, y) = x^2 - y^2$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Contour Plot of $f(x, y) = x^2 - y^2$')
plt.grid(True)
plt.show()
```

## B. Other Applications (plotting time dependent plots)

For this example, we want to expand the application of ploting contour plots to create a dynamic visualization of a 2D function that varies with time.

We also need to import `clear_output` from `IPython.display` to clear the plot for each timestep. We begin by initializing the $x$ and $y$ values and the meshgrid

```python
x_values = np.linspace(-5, 5, 100)
y_values = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x_values, y_values)
```

We then define the function that is dependent on $x$, $y$, and $t$

For this example we plotted $z=e^t\sin(x+t)\cos(y+t)$

```python
def func(x, y, t):
    return np.exp(t) * np.sin(x + t) * np.cos(y + t)
```

Lastly, we then plot the meshgrid in a `for` loop that is dependent in $t$.

We also used the `clear_output` from `IPython.display` to update the plot values.

```python
from IPython.display import clear_output
for t in np.linspace(0, 2*np.pi, 100):
    Z = func(X, Y, t)

    plt.contourf(X, Y, Z, cmap='jet', levels=20)
    plt.colorbar(label='$f(x, y, t)$')
    plt.xlabel('$x$')
    plt.ylabel('$y$')
    plt.title(f'Contour Plot of $f(x, y, t)$ for $t = {t:.2f}$')
    plt.grid(True)
    plt.pause(0.1)
    clear_output(wait=True)
    plt.close()

# Keep the plot window open until manually closed
plt.show()
```

### 6. Random Samples from an Array: [`np.random.choice()`](https://numpy.org/doc/stable/reference/random/generated/numpy.random.choice.html)

The `np.random.choice()` generates random samples from an array or a sequence. It takes the population from which to draw the samples, the number of samples to draw, and whether to sample with or without replacement.

Syntax: `numpy.random.choice(a, size=None, replace=True, p=None)`

`a`: The population array or sequence.

`size`: The number of samples to draw (can be an integer or tuple for multi-dimensional output).

`replace`: If True, sampling with replacement is allowed. If False, sampling without replacement is performed.

`p`: An optional array of probabilities associated with each entry in a. If provided, it must be the same length as a, and the elements will be normalized to sum to 1.

### B. Application

#### Random Walk
A random walk is a random process that describes a path that consists of a succession of random steps on some space.

A popular random walk model is that of a random walk on a regular lattice, where at each step the location jumps to another site according to some probability distribution. In a simple random walk, the location can only jump to neighboring sites of the lattice, forming a lattice path.

In this example we will explore a 1 Dimensional Random Walk
An elementary example of a random walk is a random walk on the integer number , which starts at $0$ and at each step moves $+1$ or $-1$ with equal probability.

First, we create a `np.random.choice`that only returns $-1$ and $1$ with $1000$ steps

```python
steps = np.random.choice([-1, 1], size=1000)
```

Then, we get the position of each steps using [`np.cumsum`](https://numpy.org/doc/stable/reference/generated/numpy.cumsum.html) which sums the value in a given axis.

```python
positions = np.cumsum(steps)
```

Lastly we plot it with the corresponding axis and labels.

```python
plt.plot(positions)
plt.xlabel('Step Number')
plt.ylabel('Position')
plt.title('Random Walk Simulation')
plt.grid(True)
plt.show()
```

Should we want to superimpose 100 of these, we can properly define a 1D random walk function.

```python
def random_walk(n):
    steps = np.random.choice([-1, 1], size=n)
    positions = np.cumsum(steps)
    return positions
```

Then we initialize the functions before plotting it in the `for` loop
- For this example we will simulate and superimpose 69 random walks of 100 steps each

```python
n = 69
num_steps = 6969
np.random.seed(696969)
plt.figure(figsize=(14, 6))

for i in range(n):
    positions = random_walk(num_steps)
    plt.plot(positions, linewidth=0.2)

plt.xlabel('Step Number')
plt.ylabel('Position')
plt.title( f'Random Walk Simulation (Superimposed {n} Walks on {num_steps} steps)')
plt.grid(True)
plt.show()
```

To get the  mean and standard deviation, we store the values on a blank array named `all_positions`. We then append the values of each position to the blank array.

We then assign a `mean_positions` and `std_positions` to determine the mean and standard deviation of each step.

Lastly, we use `plt.fill_between` to create the shade

```python
all_positions = []
plt.figure(figsize=(14, 6))
plt.grid(True)
np.random.seed(696969)

for i in range(n):
    positions = random_walk(num_steps)
    all_positions.append(positions) # We fill the all_positions with the corresponding positions for each for loop
    plt.plot(positions, linewidth=0.2)

# Calculate mean and standard deviation of positions at each step
mean_positions = np.mean(all_positions, axis=0)
std_positions = np.std(all_positions, axis=0)

# Plot the standard deviation shade
plt.fill_between(range(num_steps), mean_positions - std_positions, mean_positions + std_positions, color='gray', alpha=0.3)

plt.xlabel('Step Number')
plt.ylabel('Position')
plt.title( f'Random Walk Simulation (Superimposed {n} Walks on {num_steps} steps with Standard Deviation shade)')
plt.grid(True)
plt.show()
```

# IV. Exporting and Loading your data

### 1. Saving Numpy Arrays: [`np.save()`](https://numpy.org/doc/stable/reference/generated/numpy.save.html)

To save NumPy arrays, you can use the [`numpy.save()`](https://numpy.org/doc/stable/reference/generated/numpy.save.html) function to store the array data in a binary format with a '.npy' extension. This format is specific to NumPy and is efficient for preserving array information.

- Create a NumPy array

```python
data_array = np.array([1, 2, 3, 4, 5])
```

- Save the array to a file

```python
np.save('saved_array.npy', data_array)
```

### 2. Loading NumPy Arrays: [`np.load()`](https://numpy.org/doc/stable/reference/generated/numpy.load.html)

To load the saved NumPy array, use the `numpy.load()` function. This will read the data from the file and return it as a NumPy array.

```python
loaded_array = np.load('saved_array.npy')

print(loaded_array)
```

### 3. Saving Multiple Arrays: [`np.savez()`](https://numpy.org/doc/stable/reference/generated/numpy.savez.html)

You can also save multiple arrays into a single file using the `numpy.savez()` function. This creates a compressed '.npz' file that stores the arrays with meaningful keys.

For the example below, we'll create two arrays, namely `arr1` and `arr2`, respectively

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
```

To save multiple arrays to a single file, we still use `np.savez()`

```python
np.savez('saved_arrays.npz', arr1=arr1, arr2=arr2)
```

### 4. Loading Multiple Arrays: [`np.load()`](https://numpy.org/doc/stable/reference/generated/numpy.load.html)


To load multiple arrays from the '.npz' file, use `numpy.load()` and access the arrays using their respective keys.

```python
loaded_data = np.load('saved_arrays.npz')
loaded_array1 = loaded_data['arr1']
loaded_array2 = loaded_data['arr2']

print(loaded_array1)
print(loaded_array2)
```

### 5.Saving and Loading Text Files: [`np.savetxt()`](https://numpy.org/doc/stable/reference/generated/numpy.savetxt.html) & [`np.loadtxt()`](https://numpy.org/doc/stable/reference/generated/numpy.loadtxt.html)

NumPy also allows you to save and load arrays in plain text format using `numpy.savetxt()` and `numpy.loadtxt()` functions.

- Create and save a NumPy array

```python
data_array = np.array([1, 2, 3, 4, 5])

np.savetxt('saved_array.txt', data_array)
```

- Load the array from the text file

```python
loaded_array = np.loadtxt('saved_array.txt')

print(loaded_array)
```
